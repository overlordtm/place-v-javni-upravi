```python
import pandas as pd
import pju.fetch
```

```python
dfs = []

for m in range(1,13):
    dfs.append(pju.fetch.fetch_payouts_job_title(2021, m))
```

```python
df = pd.concat(dfs)
dfz = df[df['job_title'].str.contains('ZDRAVNIK', case=False)]
```

```python
dfz['total_salary'] = dfz.apply(lambda row: row['gross_salary'] + row['C'] + row['D'] + row['E'] + row['F'] + row['I'] + row['J'] + row['O'], axis=1)
```

```python
dfz.head()
```

```python
dfzp = pd.pivot_table(dfz, values=['gross_salary', 'total_salary', 'employees'], index='job_title', aggfunc='sum')
dfzp['nominalna_placa'] = dfzp['gross_salary'] / dfzp['employees']
dfzp['povprecna_placa'] = dfzp['total_salary'] / dfzp['employees']
dfzp['razmerje'] = dfzp['povprecna_placa'] / dfzp['nominalna_placa']
dfzp.sort_values('povprecna_placa', ascending=False)

dfzp.to_csv('zdravniki.csv')
```
