```python
import pandas as pd
import pju.fetch
```

```python
dfs = []

for m in range(1,13):
    dfs.append(pju.fetch.fetch_payouts_job_title(2022, m))
```

```python
df = pd.concat(dfs)
df.head()
```

```python
# dfz = df[df['job_title_id'].str.startswith('E01', case=False)]
dfz = df.loc[df.index.startswith('E01', case=False)]
```

```python
dfz['total_salary'] = dfz.apply(lambda row: row['gross_salary'] + row['I'] + row['J'], axis=1)
```

```python
dfz.head()
```

```python
dfzp = pd.pivot_table(dfz, values=['gross_salary', 'total_salary', 'employees', 'I', 'J'], index=['job_title', 'job_title_id'], aggfunc='sum')
dfzp['placa1'] = dfzp['gross_salary'] / dfzp['employees']
dfzp['placa2'] = dfzp['total_salary'] / dfzp['employees']
dfzp['placa_I'] = dfzp['I'] / dfzp['employees']
dfzp['placa_J'] = dfzp['J'] / dfzp['employees']
dfzp.sort_values('placa2', ascending=False)
```
